```python
import os
import pickle
import pandas as pd
import numpy as np
from src.calculate_results import (
    append_meta_data, get_all_accuracy, calculate_all_metrics
)
```

```python
ROOT_DIR = os.getcwd()
DATA_DIR = os.path.join(ROOT_DIR, "..", "data")
RESULTS_DIR = os.path.join(ROOT_DIR, "..", "results")
meta = pd.read_csv(os.path.join(DATA_DIR, "participants.tsv"), sep="\t")
```

AD vs. CN task

```python
# Load all the band results
with open(os.path.join(RESULTS_DIR, "bands/C_A/_freq_alpha/C_A/all_results.pkl"), "rb") as f:
    alpha_ad = pickle.load(f)
with open(os.path.join(RESULTS_DIR, "bands/C_A/_freq_beta/C_A/all_results.pkl"), "rb") as f:
    beta_ad = pickle.load(f)
with open(os.path.join(RESULTS_DIR, "bands/C_A/_freq_delta/C_A/all_results.pkl"), "rb") as f:
    delta_ad = pickle.load(f)
with open(os.path.join(RESULTS_DIR, "bands/C_A/_freq_gamma/C_A/all_results.pkl"), "rb") as f:
    gamma_ad = pickle.load(f)
with open(os.path.join(RESULTS_DIR, "bands/C_A/_freq_theta/C_A/all_results.pkl"), "rb") as f:
    theta_ad = pickle.load(f)
```

```python
# Append subject metadata to results
alpha_ad = append_meta_data(alpha_ad, meta)
beta_ad = append_meta_data(beta_ad, meta)
delta_ad = append_meta_data(delta_ad, meta)
gamma_ad = append_meta_data(gamma_ad, meta)
theta_ad = append_meta_data(theta_ad, meta)
```

```python
# Get test accuracies
alpha_acc = pd.DataFrame(get_all_accuracy(alpha_ad))
beta_acc = pd.DataFrame(get_all_accuracy(beta_ad))
delta_acc = pd.DataFrame(get_all_accuracy(delta_ad))
gamma_acc = pd.DataFrame(get_all_accuracy(gamma_ad))
theta_acc = pd.DataFrame(get_all_accuracy(theta_ad))
```

```python
# Calculate accuracy mean and std
print("Alpha band accuracy:", alpha_acc["test_accuracy"].mean(), "±", alpha_acc["test_accuracy"].std())
print("Beta band accuracy:", beta_acc["test_accuracy"].mean(), "±", beta_acc["test_accuracy"].std())
print("Delta band accuracy:", delta_acc["test_accuracy"].mean(), "±", delta_acc["test_accuracy"].std())
print("Gamma band accuracy:", gamma_acc["test_accuracy"].mean(), "±", gamma_acc["test_accuracy"].std())
print("Theta band accuracy:", theta_acc["test_accuracy"].mean(), "±", theta_acc["test_accuracy"].std())
```

```python
# Calculate all metrics
alpha_metrics = pd.DataFrame([calculate_all_metrics(alpha_ad)])
beta_metrics = pd.DataFrame([calculate_all_metrics(beta_ad)])
delta_metrics = pd.DataFrame([calculate_all_metrics(delta_ad)])
gamma_metrics = pd.DataFrame([calculate_all_metrics(gamma_ad)])
theta_metrics = pd.DataFrame([calculate_all_metrics(theta_ad)])
```

```python
# Concat metrics for all bands
all_band_metrics = pd.concat([
    delta_metrics,
    theta_metrics,
    alpha_metrics,
    beta_metrics,
    gamma_metrics
], axis=0, ignore_index=True)

# Add band names
all_band_metrics["band"] = ["delta", "theta", "alpha", "beta", "gamma"]
all_band_metrics
```
